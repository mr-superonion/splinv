```python
%pylab inline
import astropy.io.ascii as asciiA
```

```python
def getDataColor(data):
    mask=((data['x']>90)&(data['y']>90))
    mask=mask&((data['x']<100)&(data['y']<100))
    color=np.zeros(len(data))*0.5
    color[mask]=1.
    return color
```

```python
lbd5=asciiA.read('detection_lbd5.csv')
color5=getDataColor(lbd5)
figure=plt.figure()
ax=figure.add_subplot(1,1,1)
ax.scatter(np.log10(lbd5['M200']),lbd5['z_cl'],c=color5)
ax.set_xlim(13.5,16)
ax.set_xlabel('$log_{10}(M)$')
ax.set_ylabel('z')
```
