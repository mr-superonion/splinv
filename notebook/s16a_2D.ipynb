```python
import astropy.io.fits as pyfits
norm=mpl.colors.Normalize(vmin=-0.005,vmax=0.03)
def log_10_product(x, pos):
    """The two args are the value and tick position.
    Label ticks with the product of the exponentiation"""
    return '%d' % (x)
formatter = plt.FuncFormatter(log_10_product)
cmap=plt.cm.rainbow
fields=['HECTOMAP','VVDS','WIDE12H','GAMA15H','GAMA09H','XMM']
lEdge=50
```

```python
fig = plt.figure(figsize=(20, 12))

axes=[]

for i in range(3):
    f1=fields[i*2]
    f2=fields[i*2+1]
    xlen1=len(np.load('./raDec_%s.npy' %(f1))[0][lEdge:-lEdge])
    xlen2=len(np.load('./raDec_%s.npy' %(f2))[0][lEdge:-lEdge])
    ratio=int((xlen1/(xlen1+xlen2+0.))*100+0.5)
    axes.append(plt.subplot2grid((3, 104), (i, 0), colspan=ratio,rowspan=1))
    axes.append(plt.subplot2grid((3, 104), (i, ratio+3), colspan=100-ratio,rowspan=1))
for i,field in enumerate(fields):
    xGrid,yGrid=np.load('./raDec_%s.npy' %(field))
    xGrid=xGrid[lEdge:-lEdge]
    yGrid=yGrid[lEdge:-lEdge]
    numMap=pyfits.getdata('snrEMap_%s.fits' %(field))[lEdge:-lEdge,lEdge:-lEdge]
    ax = axes[i]
    ax.set_title(field,fontsize=15)
    im=ax.imshow(numMap,interpolation='None',\
                 extent=[xGrid[0],xGrid[-1],yGrid[-1],yGrid[0]],\
                 norm=norm,cmap='PuBu_r')
    x0,x1 = ax.get_xlim()
    y0,y1 = ax.get_ylim()
    ax.set_xlim(x1, x0)
    ax.set_ylim(y1, y0)
    #ax.set_aspect('equal', 'box')
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.95, 0.1, 0.01, 0.8])
cbar=fig.colorbar(im, cax=cbar_ax,format=formatter)#,ticks=[0, 5, 10,15,20])
plt.subplots_adjust(wspace=-0.2, hspace=-0.03,left=0.05,right=0.95,bottom=0.05,top=0.95)
#cbar.ax.set_yticklabels(['0', '5', '10' ,'15' ,'> 20'])
fig.savefig('kappaMaps.eps')
```

```python
fig=plt.figure(figsize=(10,5))
for i,eb in enumerate(['E','B']):
    ax=plt.subplot(1,2,i+1)
    for field in fields:
        a=pyfits.getdata('snr%sMap_%s.fits' %(eb,field))
        a=a[a!=0]
        print field,np.average(a),np.average(a**2.)-1,np.average(a**3.),np.average(a**4.)-3,np.average(a**5.)
        y,x=np.histogram(a,bins=40)
        x0=x[:-1]
        x1=x[1:]
        x=(x0+x1)/2.
        y= y/10000.
        ax.plot(x,y,label=field)
    ax.set_ylabel(r'num $(10^4)$',fontsize=12)
    ax.set_xlabel(r'$\kappa_%s/\sigma_%s$' %(eb,eb),fontsize=12)
plt.tight_layout()
plt.legend()
plt.savefig('PDF_kappaMap.png')
```

```python
dataAll=[]
dataAll2=[]
for field in fields:
    a=pyfits.getdata('snrEMap_%s.fits' %(field))
    a=a[a!=0]
    b=pyfits.getdata('snrBMap_%s.fits' %(field))
    b=b[b!=0]
    dataAll.append(a)
    dataAll2.append(b)
dataAll=np.hstack(dataAll)
dataAll2=np.hstack(dataAll2)
x=np.linspace(-8.,8.,20)
y=np.histogram(dataAll,bins=x,density=True)[0]
y2=np.histogram(dataAll2,bins=x,density=True)[0]
x0=x[:-1]
x1=x[1:]
x=(x0+x1)/2.
gaus=1./np.sqrt(2.*np.pi)*np.exp(-x**2./2.)
plt.plot(x,y,'o',c='red',label='E-mode')
plt.plot(x,y2,'x',c='blue',label='B-mode')
plt.plot(x,gaus,'--',c='black',label='Gaussian')
plt.yscale('log')
plt.ylim(10**-6,1.1)
plt.ylabel(r'num $(10^4)$',fontsize=12)
plt.xlabel(r'$\kappa/\sigma$',fontsize=12)
plt.legend()
plt.savefig('PDF_kappaMap.eps')
```
